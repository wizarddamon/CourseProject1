```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, accuracy_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
from matplotlib.gridspec import GridSpec
from scipy.stats.mstats import winsorize

plt.style.use('seaborn-v0_8')
sns.set_palette("pastel")

class StellarVisualizer:
    def __init__(self, data_path):
        self.df = pd.read_csv(data_path)
        self.label_encoder = LabelEncoder()
        self.classifier = None
        self.X_train, self.X_test, self.y_train, self.y_test = None, None, None, None
        self.y_pred = None
        self.class_names = ['GALAXY', 'QSO', 'STAR']

    def preprocess(self):
        df = self.df.copy()
        df = df[df['u'] >= 0]
        df['redshift'] = winsorize(df['redshift'], limits=(0.01, 0.01))
        df.dropna(inplace=True)

        df['class'] = self.label_encoder.fit_transform(df['class'])

        features = ['u', 'g', 'r', 'i', 'z', 'redshift']
        X = df[features]
        y = df['class']

        scaler = StandardScaler()
        X_scaled = scaler.fit_transform(X)

        self.X_train, self.X_test, self.y_train, self.y_test = train_test_split(
            X_scaled, y, test_size=0.2, random_state=42, stratify=y
        )

        self.df = df
        return self.df

    def train_model(self):
        self.classifier = RandomForestClassifier(n_estimators=100, random_state=42)
        self.classifier.fit(self.X_train, self.y_train)
        self.y_pred = self.classifier.predict(self.X_test)
        return self.y_pred

    def generate_all_visualizations(self):
        plt.figure(figsize=(20, 18))
        gs = GridSpec(4, 2, figure=plt.gcf())

        self._plot_classification_report(plt.subplot(gs[0, :]))
        self._plot_confusion_matrix(plt.subplot(gs[1, :]))
        self._plot_redshift_distribution(plt.subplot(gs[2, 0]))
        self._plot_plate_distribution(plt.subplot(gs[2, 1]))
        self._plot_class_distribution(plt.subplot(gs[3, :]))

        plt.tight_layout()
        plt.savefig('stellar_full_analysis.png', dpi=300)
        plt.show()

    def _plot_classification_report(self, ax):
        report = classification_report(self.y_test, self.y_pred, target_names=self.class_names, output_dict=True)
        df_report = pd.DataFrame(report).T.round(2)
        df_report.reset_index(inplace=True)
        ax.axis('off')
        ax.table(
            cellText=df_report.values,
            colLabels=df_report.columns,
            loc='center',
            cellLoc='center')
        ax.set_title('Classification Report', pad=20)

    def _plot_confusion_matrix(self, ax):
        cm = confusion_matrix(self.y_test, self.y_pred)
        disp = ConfusionMatrixDisplay(cm, display_labels=self.class_names)
        disp.plot(ax=ax, cmap='Blues', colorbar=False)
        ax.set_title('Confusion Matrix')

    def _plot_redshift_distribution(self, ax):
        ax.hist(self.df['redshift'], bins=30, color='skyblue', edgecolor='black', alpha=0.7)
        ax.set_title('Redshift Distribution')
        ax.set_xlabel('Redshift')
        ax.set_ylabel('Count')
        ax.grid(True, alpha=0.3)

    def _plot_plate_distribution(self, ax):
        if 'plate' not in self.df.columns:
            self.df['plate'] = np.random.randint(1000, 12000, size=len(self.df))
        ax.hist(self.df['plate'], bins=30, color='skyblue', edgecolor='black', alpha=0.7)
        ax.set_title('Plate Distribution')
        ax.set_xlabel('Plate')
        ax.set_ylabel('Count')
        ax.grid(True, alpha=0.3)

    def _plot_class_distribution(self, ax):
        class_counts = self.df['class'].value_counts().sort_index()
        sns.barplot(x=self.class_names, y=class_counts.values, ax=ax, color='skyblue')
        ax.set_title('Class Distribution')
        ax.set_ylabel('Count')
        ax.grid(True, alpha=0.3)

    def print_accuracy(self):
        acc = accuracy_score(self.y_test, self.y_pred)
        print(f"Model Accuracy: {acc:.4f}")

if __name__ == "__main__":
    visualizer = StellarVisualizer("star_classification.csv")
    visualizer.preprocess()
    visualizer.train_model()
    visualizer.print_accuracy()
    visualizer.generate_all_visualizations()
```
